```python
# pandas 라이브러리 - 데이터 분석하기

# DataFrame을 생성하기 위한 기본 데이터
data = {
  '이름' : ['김자바', '이자바', '박자반', '최자바', '정자바'],
  '학급' : ['자바반', '자바반', '자바반', '캐드반', '캐드반'],
  '국어점수' : [90, 80, 70, 100, 90],
  '영어점수' : [70, 80, 80, 70, 90],
  '평균' : [80.0, 80.0, 75.0, 85.0, 90.0],
  '취미' : ['독서', '', '', '요가', '축구']
}

```

```python
import pandas
```

```python
df = pandas.DataFrame(data)
df 
```

```python
#데이터프레임을 엑셀파일로 저장
df.to_excel('text1.xlsx')

# 에러 ModuleNotFoundError: No module named 'openpyxl(오류뜨는 모듈을 입력하셋)' 이거 뜨면 설치 안되어있다고 알려주는거
# 터미널 open해서 >> pip install [(오류뜨는 모듈을 입력하셋] 호출하기.
```
